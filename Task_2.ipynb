```python
df['Genre'].value_counts()
```

```python
bars =df['Genre'].value_counts()
bars.plot(kind="bar",title='Top Genres',figsize=(8,5),color='yellow')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Year",data= df, order = df.groupby(by=['Year'])['Review'].count().sort_values(ascending=False).index)
plt.xticks(rotation='horizontal')
plt.title('Games released')                                                   # adding a title to the data
plt.ylabel('Count')                                                           # labelling the y-axis by count.
plt.xlabel('Year')                    # labelling the x-axis by year.
```

```python
gradable_counts=df.Platform.value_counts().nlargest(10)
gradable_counts.plot(kind='pie', title='Top 10 Platforms', figsize=(8,8),autopct='%.2f %%')
plt.legend()
plt.show()
```

```python
genre = df.Genre.value_counts().values
labels =df.Genre.value_counts().index
plt.title('Genre by Sales')
plt.pie(genre,labels=labels,autopct='%.2f %%',pctdistance=0.9)
plt.legend(bbox_to_anchor=(1.5,1.0),loc='upper left')
plt.show()
```

```python
platform = df.Platform.value_counts()
plt.figure(figsize=(12,8))
plt.barh(platform.index,platform.values)
for i,v in enumerate(platform.values):
    plt.text(v,i,str(v))
plt.title('Sales by Platform',)
plt.xlabel('Counts')
plt.ylabel("Platform")
plt.show()
```
